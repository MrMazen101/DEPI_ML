# <a id="1"> 1 ) . What is the distribution for User type?</a>

```python
## I will import seaborn library to support me in making visualizations 

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
base_color = sns.color_palette()[0]

sns.countplot(data=df, x='user_type', color=base_color)

plt.xlabel('User Type')
plt.ylabel('Frequency')
plt.title("Distribution of User Types")
plt.show()

#
```

> ### <mark>Your analsys and exploration here </mark>


The visualization shows the distribution of users based on whether they are part of the bike share program or not. The majority of users are not part of the bike share program, indicating that most users are casual riders rather than members. This insight is valuable for understanding user behavior and can inform marketing strategies to encourage more users to join the bike share program.

# <a id="2"> 2 ) What is the distribution for the bike share users on all Trips?</a>

```python
plt.figure(figsize = (10,5), dpi = 200) 
sns.countplot(data=df, x='bike_share_for_all_trip')
plt.xlabel('User distribution')
plt.ylabel('Frequency')
plt.title("Representing the distribution of user_type")
plt.show()
```

> ### <mark>Your analsys and exploration here </mark>

The overwhelming majority of users are not enrolled in the "Bike Share for All Trips" program. Only a very small percentage of users participate in this specific program.
