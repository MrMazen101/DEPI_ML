### Method 1: Drop Columns

### Method 3 
Mean/Median: Used for numerical data when you want to keep the rows. Use Mean for normal distribution and Median if there are outliers .

Mode: Used for categorical data (text) to replace missing values with the most common category

```python

df_imputed = df.copy()
```

```python
numerical_cols = df_imputed.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = df_imputed.select_dtypes(include=['object']).columns
```

```python
for col in numerical_cols:
    mean_value = df_imputed[col].mean()
    df_imputed[col].fillna(mean_value, inplace=True)
    
```

```python
for col in categorical_cols:
   if df_imputed[col].isnull().sum() > 0:
        mode_value = df_imputed[col].mode()[0]
        df_imputed[col].fillna(mode_value, inplace=True)
```

```python
print("missing values after imputation:\n", df_imputed.isnull().sum())
```
