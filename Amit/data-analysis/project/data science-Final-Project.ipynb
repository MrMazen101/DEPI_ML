
# <a id="13">  13 ) . What is the  Correlation between each variable in the dataset?</a>

```python
# Make visualization to easy get insight
corr_matrix = df.select_dtypes(include=['number']).corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Variables')
plt.show()

```


# <a id="14"> 14 )  .  What is the  Correlation between `specific numeric variables` in the dataset?</a>


```python
# Make visualization to easy get insight
plt.figure(figsize=(8, 5))
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap='vlag', fmt='.2f', center=0)
plt.title('Correlation Between Variables in the Dataset')
plt.show()

```

> ### <mark>Your analsys and exploration here </mark>

Example usage
- There is `no Correlation` between the user's `age` and the trip `duration` (whether in seconds, minutes, or hours).
- There is a `perfect positive correlation` (1.0) between `duration_sec`, `duration_min`, and `duration_hour` since they represent the exact same metric.
- There are no strong relationships between any of the numeric variables in this dataset other than the time conversions.



# <a id="15">  15 ) What is the  Correlation between data_stat variables in the dataset?</a>


```python
plt.figure(figsize=(8, 5))
numeric_vars = ['age', 'duration_min', 'duration_hour']
corr_matrix = df[numeric_vars].corr()
sns.heatmap(corr_matrix, annot=True, cmap='Blues', fmt='.3f', center=0)
plt.title('Correlation Between Key Numeric Variables (Age & Duration)')
plt.show()
```
