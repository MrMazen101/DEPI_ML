```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import dbscan_lab_helper as helper
```

```python
df = pd.read_csv('Blobs.csv')
data_1 = df.values
```

```python
helper.plot_dataset(data_1)
```

```python
from sklearn import cluster
dbscan = cluster.DBSCAN(eps=0.5, min_samples=5)
cluster_labels = dbscan.fit_predict(data_1)
```

```python
helper.plot_clustered_dataset(data_1, cluster_labels)
```

```python
epsilon = 2
dbscan = cluster.DBSCAN(eps=epsilon, min_samples=5)
cluster_labels = dbscan.fit_predict(data_1)
helper.plot_clustered_dataset(data_1, cluster_labels)
```
