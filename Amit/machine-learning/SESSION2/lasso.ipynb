```python
import numpy as np
```

```python
class LinearRegression:
    #ده بيكون الكونستركتور بتاع الكلاس اللي بيتم استدعاؤه لما بنعمل اوبجكت من الكلاس    
    def __init__(self, alpha=0.01, n_iters=1000):
        self.alpha = alpha
        self.n_iters = n_iters
        self.w = None
        self.b = None
    
#هنا بنعمل فيت للبيانات اللي عندنا عشان نطلع الوي والبي اللي هما الباراميترز بتوع الموديل   
    def fit(self, X, y):
        n_samples, n_features = X.shape
        # init parameters
        self.w = np.zeros(n_features)
        self.b = 0

        # gradient descent
        for _ in range(self.n_iters):
            y_predicted = np.dot(X, self.w) + self.b

            dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y))
            db = (1 / n_samples) * np.sum(y_predicted - y)

            self.w -= self.alpha * dw
            self.b -= self.alpha * db
    #هنا بنعمل فيت للبيانات اللي عندنا عشان نطلع الوي والبي اللي هما الباراميترز بتوع الموديل مع اضافة الريجولاريزيشن اللي هو لامبدا 
    def fit_ridge(self, X, y, lambda_):
        n_samples, n_features = X.shape
        # init parameters
        self.w = np.zeros(n_features)
        self.b = 0

        # gradient descent
        for _ in range(self.n_iters):
            y_predicted = np.dot(X, self.w) + self.b

            dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y)) + (lambda_ / n_samples) * self.w
            db = (1 / n_samples) * np.sum(y_predicted - y)

            self.w -= self.alpha * dw
    
            self.b -= self.alpha * db
    #هنا بنعمل فيت للبيانات اللي عندنا عشان نطلع الوي والبي اللي هما الباراميترز بتوع الموديل مع اضافة الريجولاريزيشن اللي هو لامبدا مع لاسو اللي هو بيعمل سبارسيتى للويز يعني بيخلي بعض الويز بتوعنا صفر عشان يختار اهم الفيتشرز اللي بتاثر على البريديكشن بتاعتنا
    def fit_lasso(self, X, y, lambda_):
        n_samples, n_features = X.shape
        # init parameters
        self.w = np.zeros(n_features)
        self.b = 0

        # gradient descent
        for _ in range(self.n_iters):
            y_predicted = np.dot(X, self.w) + self.b

            dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y)) + (lambda_ / n_samples) * np.sign(self.w)
            db = (1 / n_samples) * np.sum(y_predicted - y)

            self.w -= self.alpha * dw
            self.b -= self.alpha * db
    #هنا بنعمل بريديكشن للبيانات اللي عندنا باستخدام الوي والبي اللي احنا طلعناهم من الفيت  
    def predict(self, X):
        linear_model = np.dot(X, self.w) + self.b
        y_predicted = linear_model
        return y_predicted
    
```
