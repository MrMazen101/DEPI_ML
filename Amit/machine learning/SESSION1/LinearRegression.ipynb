```python
import numpy as np
```

```python
class LinearRegression_Mohamed_Hamada:
    def __init__(self ,lr =0.01 ,iteration =10 ,w=1 ,b=1):
        self.lr = lr
        self.iteration = iteration
        self.w = w
        self.b = b

    def fit(self,x,y):
        for _ in range(self.iteration):

            y_pred = self.w * x +self.b

            loss = (2/x.shape[0])* np.sum(y_pred - y)

            dw = (2/x.shape[0])* np.sum((y_pred - y)*x)
            db = (2/x.shape[0])* np.sum(y_pred - y)

            self.w = self.w - self.lr * dw
            self.b = self.b - self.lr *db
        
        return self.w ,self.b 

    def predict(self,x):
        return self.w *x +self.b
    
    
```

```python

X = np.array([50, 60, 70, 80, 90])
y = np.array([150, 180, 210, 240, 270])

reg = LinearRegression_Mohamed_Hamada(0.0001 ,100 ,0 ,0)

new_w ,new_b = reg.fit(X ,y)

print("w:", new_w)
print("b:", new_b)
print("Predictions:", reg.predict(X))
```
