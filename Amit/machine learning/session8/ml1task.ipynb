```python
import numpy
```

```python
class linearRegression_mazenyehia:
    def __init__(self, learning_rate=0.01, n_iters=1000):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n=float(len(X))
        

        # gradient descent
        for _ in range(self.n_iters):
            y_predicted = self.weights * X + self.bias
            dw = (2 / n) * numpy.sum((y_predicted - y)*X)
            db = (2 / n) * numpy.sum(y_predicted - y)

            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db
            print(f"weights: {self.weights}, bias: {self.bias}")    
    def predict(self, X):
        y_approximated = self.weights * X + self.bias
        return y_approximated       
        
```

```python
X_train = numpy.array([1, 2, 3, 4])
Y_train = numpy.array([2, 2.8, 3.6, 4.5])
```

```python
model = linearRegression_mazenyehia(learning_rate=0.01, n_iters=2)
model.weights = 0
model.bias = 0
model.fit(X_train, Y_train)
```

```python
print(f"\nFinal parameters: w = {model.weights}, b = {model.bias}")
```
