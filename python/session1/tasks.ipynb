## üëçtask one it's Suppose you have this email address ‚ÄúAmit_ml@gmail.edu‚Äù

```python
def sign(email) :
    if email.count('@')!=1:
        return "invalid email"
    
    index = email.find ('@')
    if '.'not in email [index:]:
          return "invalid"
    
    username = email[:index]
    dot_index = email.rfind('.')
    d_name = email[index+1 : dot_index ]

    do_type =""
    if email.endswith ('.com'):
         do_type="commercial domain1"
    elif email.endswidth ('.edu'):
         do_type ="educational domain"
    else :
         do_type ="other domain"
      
    
    return {
         "username": username,
         "domain" : d_name,
         "type": do_type
    }

result = sign("Amit_ml@gmail.edu")
print(result)
```

## task2_decode

```python
def decode():
    encode_message = "###!!@emocleW EPGTQ!!!6789"
    clean_message = encode_message.strip("#!@ 0123456789")
    words = clean_message.split()
    word1 = words[0] 
    word2 = words[1] 
    decode_word1 = word1[::-1] 
    decode_word2 = word2 
    print(f"Result: {decode_word1} {decode_word2}")


decode()
```

## task 3

```python
def decode_1():
    encode_message = "&&&**$gnirtS PLIO!!81234"
    clean_message = encode_message.strip("&*$! 0123456789")
    words = clean_message.split()
    word1 = words[0] 
    word2 = words[1] 
    decode_word1 = word1[::-1] 
    decode_word2 = word2.replace('I', 'E').replace('O', 'U') 
    print(f"Task 3 Result: {decode_word1} {decode_word2}")


decode_1()
```
